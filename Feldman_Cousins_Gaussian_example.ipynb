```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 0.5
x=np.linspace(-5,5,1001)  ## has to be odd for the code to work, need zero

P =  np.exp(-0.5*(x-mu)**2)
P /=np.sum(P)

plt.plot(x,P)
plt.title("gaussian probability distribtion")
plt.xlabel("x")
plt.ylabel("$\mu$")
plt.show()
```

```python
def computeR(x,P):
    P_best_xgt0 = np.max(P)
    P_best_xlt0 = np.exp(-1/2 * x**2)
    P_best_xlt0 /= np.sum(P_best_xlt0)
    
    print(P_best_xgt0,P_best_xlt0,(np.where(np.abs(x) == np.min(np.abs(x))))[0])
    
    R = np.zeros(np.shape(x))
    pos = np.where(x>0)
    neg = np.where(x<=0)
    
    R[pos] = P[pos] / P_best_xgt0
    R[neg] = P[neg] / P_best_xlt0[neg]
    return(R)

R = computeR(x,P)
    
plt.plot(x,R)
plt.title("Feldman Cousins R")
plt.xlabel("x")
plt.ylabel("$\mu$")
plt.show()
```

```python
in_integral = np.where(R>.39)
print("Integrated Prob: ",np.sum(P[in_integral]),
        "x_min: ", x[np.min(in_integral)],
      "x_max: ", x[np.max(in_integral)] )
```
