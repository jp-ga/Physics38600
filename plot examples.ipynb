```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
```

```python
x = np.linspace(50,200,1000)
y = np.ones(np.shape(x))*.985
y+=0.01*np.sin(x*2*np.pi/13.)
y+=0.01*np.sin(x*2*np.pi/23.)
y[np.where(np.logical_or(x<80,x>160))] = 0

ga= np.exp(-.5*((x-125)/3.)**2)

y = np.fft.fftshift(np.fft.ifft(np.fft.fft((ga))*np.fft.fft((y))))
y /= np.max(y)*1.01
y = np.real(y)

plt.plot(x,y)
plt.title("makes bandwidht look narrow and conveys complication")
plt.ylim(0.97,1)
#plt.xlim(70,170)
plt.ylabel("transmission [Arb.]")
plt.xlabel("frequency [GHz]")
plt.show()
```

```python
ax = plt.subplot(111)

plt.plot(x,y*100,linewidth=3,c="royalblue")
#plt.title("makes bandwidht look wide and conveys how high the coupling is")
plt.ylim(0,1)
plt.xlim(70,170)


plt.xticks([80,120,160])
plt.yticks([0,100],labels=["0%","100%"])

#ax.spines[[ 'right']].set_visible(False)

axis_font = {'fontname':'Arial', 'size':'20'}
plt.ylabel("Transmission", **axis_font)
plt.xlabel("Frequency [GHz]", **axis_font)
plt.show()
```

## simplicity examples

```python
f1 = x**4 *np.exp(-(x/180.)**20)*np.sin(x)
f1/=np.max(f1)/1e-15
f2 = x**6 *np.exp(-(x/180.)**15)*np.sin(x*2.7)
f2/=np.max(f2)/1e-15/3

plt.plot(x,f1,label='simulation 1')
plt.plot(x,f2,"orange",label='simulation 2')


#plt.title("makes bandwidht look wide and conveys how high the coupling is")
#plt.ylim(0,1)
#plt.xlim(70,170)
plt.grid(fillstyle="full",linewidth = "1.4", linestyle = "-.")
plt.legend()
plt.ylabel("strain amlitude [Arb.]")
plt.xlabel("time [ms]")
plt.show()
```

```python
ax = plt.subplot(111)
ax.plot(x-200,f2*1e15,"limegreen",linewidth=.5)
ax.plot(x-200,f1*1e15,"b",linewidth=1.5)

title_font = {'fontname':'Arial', 'size':'16', 'color':'black', 'weight':'normal',
              'verticalalignment':'bottom'} # Bottom vertical alignment for more space
axis_font = {'fontname':'Arial', 'size':'16'}




plt.ylabel("Gravitational Strain [arb]", **axis_font)
plt.xlabel("Time to Merger [ms]", **axis_font)
plt.xticks([-150,-100,-50, -0])
plt.yticks([-3, 0, 3])
ax.spines[['right', 'top']].set_visible(False)
ax.xaxis.set_tick_params(width=1.)
ax.yaxis.set_tick_params(width=1.)
plt.text(-145, -0.5, r"$2M_\odot+3M_\odot \rightarrow 4 M_\odot $", c="b")
plt.text(-87, -2.5, r"$4M_\odot+5M_\odot \rightarrow 7 M_\odot $", 
             c="limegreen",rotation=-37)

plt.show()
```
